### Path Type: Straight Path

```python
relative_path = []
for y_pattern in xrange(0, 7, 3): 
    for x in xrange(6, 42, 4):
        for z in xrange(2, 48):
            for y in range(5 + y_pattern,7 + y_pattern):
                relative_path.append((x,z,y))
#        for z in xrange(3, 48, 3): 
#            for x_2 in xrange(x, x + 6): 
#                for y in range(6 + y_pattern,7 + y_pattern):
#                    relative_path.append((x_2,z,y))
#            for x_3 in xrange(x - 5, x): 
#                for y in range(6 + y_pattern,7 + y_pattern):
#                    relative_path.append((x_3,z,y))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

try:
    xs = [i[0] for i in relative_path]
    zs = [j[1] for j in relative_path]
    ys = [k[2] for k in relative_path]
    ax.scatter(xs, zs, ys, color='green', marker='|')
    ax.set_zlim3d([0, 33])
    ax.set_xlabel('X Label')
    ax.set_ylabel('Z Label')
    ax.set_zlabel('Y Label')
except KeyError:
    pass
```
