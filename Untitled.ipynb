### MinePath Eval Optimize across various variabls

```python
world = block_handler.load_world('/Users/ryanlambert/minecraft-server/world')
```

```python
one_by_two_efficiency = []
for x_spacing in range(3, 4):
    one_by_two = []
    for y_pattern in xrange(0, 4, 3): 
        for x in xrange(6, 42, x_spacing):
            for z in xrange(2, 48):
                for y in range(6 + y_pattern,8 + y_pattern):
                    one_by_two.append((x,z,y))

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    try:
        xs = [i[0] for i in one_by_two]
        zs = [j[1] for j in one_by_two]
        ys = [k[2] for k in one_by_two]
        ax.scatter(xs, zs, ys, color='green', marker='|')
        ax.set_zlim3d([0, 17])
        ax.set_xlabel('X Label')
        ax.set_ylabel('Z Label')
        ax.set_zlabel('Y Label')
    except KeyError:
        pass

    one_by_two_ore_filter = {
        'iron':0,
        'diamond':0,
        'redstone':0,
    #    'coal':0,
        'gold':0,
    }

    blocks, one_by_two_absolute_path = block_handler.simulation(
        x1=-10,
        x2=10,
        z1=-10,
        z2=10,
        y1=0,
        y2=1,
        relative_path=one_by_two,
        ore_count=one_by_two_ore_filter,
        world=world)

ore_counts = pd.DataFrame([one_by_two_ore_filter], 
                          index=['one_by_two'])
one_by_two_efficiency = ore_counts / len(one_by_two_absolute_path) * 100
one_by_two_efficiency
```
